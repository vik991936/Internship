### 1) Write a python program to display all the header tags from wikipedia.org.

```python
import  requests                #import the libraries requests and BeautifulSoup
from  bs4 import BeautifulSoup

url_link = 'https://en.wikipedia.org/wiki/Main_Page'         #pass a URL into a variable

request = requests.get(url_link)             #use the request library to fetch the URL

soup = BeautifulSoup(request.content,'html.parser')        #create a BeautifulSoup object

heading = ['h1','h2','h3','h4','h5','h6']            # creating a list of all common heading

for i in soup.find_all(heading):            #iterate over all the heading using find_all() method
    print(i.name + ' - ' + i.text.strip())
```

### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup


# Creating the lists we want to write into
titles = []
years = []
imdb_ratings = []
ranking = []

pages = np.arange(1, 101, 50)
pages

# Storing each of the urls of 50 movies 
for page in pages:
    # Getting the contents from the each url
    page = requests.get('https://www.imdb.com/search/title/?groups=top_1000&start=' + str(page) + '&ref_=adv_nxt')
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # Aiming the part of the html we want to get the information from
    movie_div = soup.find_all('div', class_='lister-item mode-advanced')
    

    
    for movie in movie_div:
        # Scraping the movie's name
        name = movie.h3.a.text
        titles.append(name)
        
        # Scraping the movie's rank
        rank = movie.h3.find('span', class_='lister-item-index unbold text-primary').text
        ranking.append(rank)
        
        # Scraping the movie's year
        year = movie.h3.find('span', class_='lister-item-year').text.strip('()')
        years.append(year)
        
        # Scraping the rating
        imdb = float(movie.strong.text)
        imdb_ratings.append(imdb)
        
movies = pd.DataFrame({'Movie':titles,
                       '   Year':years,
                       'IMDB_rating':imdb_ratings,
                      'Rank of movie':ranking})

movies.head()

```

### 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.


```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup


# Creating the lists we want to write into
titles = []
ranking = []
years = []
imdb_ratings = []


# Getting the contents from the url

page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/?sort=')
soup = BeautifulSoup(page.text, 'html.parser')
    
# Aiming the part of the html we want to get the information from
movies = soup.find('tbody', class_='lister-list').find_all("tr")
       
for movie in movies:
    # Scraping the movie's rank 
    rank = movie.find('td', class_="titleColumn").get_text(strip=True).split('.')[0]
    ranking.append(rank)
    
    # Scraping the movie's name
    name = movie.find('td', class_="titleColumn").a.text
    titles.append(name)
    
    # Scraping the movie's year
    year = movie.find('td', class_="titleColumn").span.text.strip('()')
    years.append(year)
    
    # Scraping the rating
    imdb = float(movie.find('td', class_='ratingColumn imdbRating').strong.text)
    imdb_ratings.append(imdb)
    
    
_movies = pd.DataFrame({'Movie Rank':ranking,
                        'Movie Name':titles,
                       'release year of movie':years,
                       'IMDB_rating':imdb_ratings})

_movies
```

### 4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp .

```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup


# Creating the lists we want to write into
product_name = []
price = []
discounts = []

# Getting the contents from the url
page = requests.get('https://meesho.com/bags%02ladies/pl/p7vbp')
soup = BeautifulSoup(page.text, 'lxml')
    
# Aiming the part of the html we want to get the information from
    
all_product = soup.find('div', attrs= {'class':"sc-gsDKAQ Grid__Row-sc-4ki5nk-0 iQiJjX gfsqfL products"}).find_all('div', attrs = {'class':"sc-dkPtRN ProductList__GridCol-sc-8lnc8o-0 FjWWx jMkQHN"})


for product in all_product:
        
    #Scraping the product's name
    name = product.find('a', href=True).div.find('div',attrs = {'class':"Card__BaseCard-sc-b3n78k-0 dUNFgg NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS"})
    
    names = name.p.text
    product_name.append(names)
    
       
    #Scraping the product's price
    prices = product.find('div', attrs = {'class':"Card__BaseCard-sc-b3n78k-0 iLPHgK NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr"}).p.text
    price.append(prices)
     
    #Scraping the discount on products
    dis = product.find('div', attrs = {'class':"Card__BaseCard-sc-b3n78k-0 iLPHgK NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr"}).span.text
    discounts.append(dis)
    

bags = pd.DataFrame({'Product Name': product_name,
                    'Original Price': price,
                     'Discount': discounts})
bags


```

### 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
    

#####       a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup


# Creating the lists we want to write into
team_rank = []
team_name = []
matches = []
points = []
rating = []

# Getting the contents from the url
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup = BeautifulSoup(page.text, 'lxml')


top_teams = soup.tbody.find_all('tr')

for teams in top_teams:
    
    #scraping of teams rank.
    rank = teams.td.text
    team_rank.append(rank)
    
    #scraping of teams name.
    name = teams.find('td', class_="rankings-block__banner--team-name").text
    team_name.append(name)

    #scraping of matches.
    mat = teams.find('td', class_="rankings-block__banner--matches").text
    matches.append(mat)
    
    #scraping of points.
    point = teams.find('td', class_="rankings-block__banner--points").text
    points.append(point)
    
    #scraping of rating.
    rate = teams.find('td', class_="rankings-block__banner--rating u-text-right").text
    rating.append(rate)
    
    
        
rank_of_teams = pd.DataFrame({'Team rank': team_rank,
                    'Team name': team_name,
                     'Matches': matches,
                    'Points': points,
                    'Rating': rating})
rank_of_teams
```
